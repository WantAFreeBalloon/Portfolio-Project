```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']= (12,8) # adjusts the configuration of the plot we will create


# Read in the data
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
df.isnull().sum()
len(df)
```

```python
df.dtypes
```

```python
budget_avg = df.budget.mean()
df.budget.fillna(value = budget_avg, inplace = True)

gross_avg = df.gross.mean()
df.gross.fillna(value = gross_avg, inplace = True)
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head(15)
```

```python
df[['Correct_Month', 'B']] = df['released'].str.split(' ', 1, expand=True)
df[['Correct_Date', 'Correct_Year']] = df['B'].str.split(',', 1, expand=True)

df.drop('B', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Correct_Year']  = df['Correct_Year'].astype(str).str[1:5]
```

```python
df.head()
```

```python
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
#pd.set_option('display.max_rows', None)
```

```python
# drop any duplicates
df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
#drop any duplicates
df.drop_duplicates()
```

```python
df
```

```python
# Scatter Plot: budget vs gross revenue

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()



```

```python
# Plot budget vs gross using seaborn
sns.regplot(x='budget', y = 'gross', data = df, scatter_kws = {"color":"red"},line_kws = {'color':'blue'})
```

```python
# Correlation
df.corr(method = 'pearson') 
#pearson, kendall, spearman
```

```python
correlation_matrix = df.corr(method = 'pearson') 

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matric for Numeric Featuring')

plt.xlabel('Movie Features')

plt.ylabel('Movies Features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype =='object'):
         df_numerized[col_name] = df_numerized[col_name].astype('category')
         df_numerized[col_name]= df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
